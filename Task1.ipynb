```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
t=pd.read_csv("Employee Attrition.csv")
t.head
print(t.isnull())
```

```python
a=t.isnull().sum()
t.dropna()
```

```python
total=np.product(t.shape)
t_missing=a.sum()
(t_missing/total)*100
```

```python
t.fillna(method='bfill',axis=0).fillna(0)
```

```python
t['satisfaction_level'].value_counts().plot(kind='bar',color=['#d5e0fe','#656371','#ff7369'])
```

```python
sns.set_style('dark')
sns.scatterplot(x=t.time_spend_company,y=t.satisfaction_level,hue=t.number_project)
```

```python
correlation_matrix=t.corr()
```

```python
correlation_matrix
```

```python
sns.set_style('dark')
sns.heatmap(correlation_matrix,annot=True,linecolor='white',linewidths=5,cmap="YlGnBu")
```
